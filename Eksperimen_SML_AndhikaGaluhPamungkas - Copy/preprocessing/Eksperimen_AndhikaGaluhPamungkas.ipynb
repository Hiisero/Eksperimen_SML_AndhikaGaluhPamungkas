Jika Anda menggunakan data teks, data mentah sering kali mengandung nilai kosong, duplikasi, atau rentang nilai yang tidak konsisten, yang dapat memengaruhi kinerja model. Oleh karena itu, proses ini bertujuan untuk membersihkan dan mempersiapkan data agar analisis berjalan optimal.

Cukup sesuaikan dengan karakteristik data yang kamu gunakan yah. Khususnya ketika kami menggunakan data tidak terstruktur.

```python
# Memisahkan fitur dan label
X = df.drop('target', axis=1)
y = df['target']

X.shape, y.shape
```

```python
# Inisialisasi scaler
scaler = StandardScaler()

# Melakukan standarisasi fitur
X_scaled = scaler.fit_transform(X)

# Mengubah hasil standarisasi ke DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

X_scaled_df.head()
```

Standarisasi data dilakukan menggunakan metode StandardScaler untuk mengubah skala setiap fitur sehingga memiliki nilai rata-rata 0 dan standar deviasi 1. Proses ini bertujuan agar seluruh fitur memiliki skala yang sebanding, sehingga model machine learning tidak bias terhadap fitur dengan rentang nilai yang lebih besar.

```python
# Menggabungkan fitur yang telah distandarisasi dengan label
processed_df = X_scaled_df.copy()
processed_df['target'] = y.values

processed_df.head()
```

Fitur yang telah diproses digabungkan kembali dengan label target.

```python
import os

output_dir = "preprocessing"
os.makedirs(output_dir, exist_ok=True)

# Menyimpan dataset hasil preprocessing
processed_df.to_csv(
    f"{output_dir}/heart_preprocessing.csv",
    index=False
)
```
